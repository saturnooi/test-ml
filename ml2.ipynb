```python
from PIL import Image
import numpy as np
import os

def load_images(folder, label):
    images = []
    labels = []
    for filename in os.listdir(folder):
        if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"): 
            img = Image.open(os.path.join(folder, filename))
            img = img.resize((128, 128))  # ปรับขนาดภาพ
            img = img.convert('RGB')
            img_array = np.array(img)
            images.append(img_array)
            labels.append(label)
    return images, labels
```

```python
from tensorflow.python.keras.models import load_model
model = load_model('/Users/nuttawatjanwisit/Desktop/nut/ml/base_model.h5')
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# โหลดข้อมูล
images_like, labels_like = load_images('/Users/nuttawatjanwisit/Desktop/nut/ml/ex-img', 1)
images_dislike, labels_dislike = load_images('/Users/nuttawatjanwisit/Desktop/nut/ml/ex-img2', 0)

# รวมข้อมูลและแปลงเป็น numpy array
all_images = np.array(images_like + images_dislike)
all_labels = np.array(labels_like + labels_dislike)

# ฝึกโมเดล
model.fit(all_images, all_labels, epochs=1000)
```

```python
def predict_image(file_path):
    img = Image.open(file_path)
    img = img.resize((128, 128))
    img = img.convert('RGB')
    img_array = np.array(img)
    img_array = np.expand_dims(img_array, axis=0)
    prediction = model.predict(img_array)
    return prediction

# ตัวอย่างการใช้งาน
prediction = predict_image('/Users/nuttawatjanwisit/Desktop/nut/ml/test/S__13946625.jpg')
if prediction[0][0] > prediction[0][1]:
    print("ไม่ชอบ")
else:
    print("ชอบ")
```

```python
import os
import shutil
from PIL import Image
import numpy as np

def predict_images(folder_path, like_folder, not_like_folder):
    # List all files in the folder
    files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

    # Create directories for 'like' and 'not like' if they don't exist
    os.makedirs(like_folder, exist_ok=True)
    os.makedirs(not_like_folder, exist_ok=True)

    # Loop through each file
    for file in files:
        # Construct full file path
        file_path = os.path.join(folder_path, file)

        # Open the image
        img = Image.open(file_path)
        img = img.resize((128, 128))
        img = img.convert('RGB')

        # Convert the image to a numpy array and add a batch dimension
        img_array = np.array(img)
        img_array = np.expand_dims(img_array, axis=0)

        # Make a prediction
        prediction = model.predict(img_array)

        # Move the file based on the prediction result
        if prediction[0][0] > prediction[0][1]:
            shutil.move(file_path, os.path.join(not_like_folder, file))
            print(f"{file}: ไม่ชอบ, moved to {not_like_folder}")
        else:
            shutil.move(file_path, os.path.join(like_folder, file))
            print(f"{file}: ชอบ, moved to {like_folder}")

# Example usage
predict_images('/Users/nuttawatjanwisit/Desktop/nut/ml/test', 
               '/Users/nuttawatjanwisit/Desktop/nut/ml/test-like', 
               '/Users/nuttawatjanwisit/Desktop/nut/ml/test-dot-like')
```

```python
import os
import requests
from PIL import Image
from io import BytesIO
import numpy as np

def download_image(url):
    response = requests.get(url)
    if response.status_code == 200:
        img = Image.open(BytesIO(response.content))
        return img
    else:
        return None

def clean_filename(url):
    # Extracting the base name and replacing problematic characters
    filename = os.path.basename(url).split('?')[0]  # Remove query parameters
    if not filename.lower().endswith(('.png', '.jpg', '.jpeg', '.webp', '.gif')):
        filename += '.jpg'  # Defaulting to .jpg if no valid extension is found
    return filename

def predict_and_save_images(file_path, like_folder, not_like_folder):
    # Ensure the like and not-like folders exist
    if not os.path.exists(like_folder):
        os.makedirs(like_folder)
    if not os.path.exists(not_like_folder):
        os.makedirs(not_like_folder)

    # Read URLs from the file
    with open(file_path, 'r') as file:
        urls = file.readlines()

    # Process each URL
        print(urls)
    for url in urls:
        url = url.strip()
        if url:
            print(url)
            img = download_image(url)
            if img is not None:
                img = img.resize((128, 128))
                img = img.convert('RGB')
                img_array = np.array(img)
                img_array = np.expand_dims(img_array, axis=0)
                prediction = model.predict(img_array)
                
                # Clean the filename and save the image in the appropriate folder
                clean_name = clean_filename(url)
                if prediction[0][0] > prediction[0][1]:
                    img.save(os.path.join(not_like_folder, clean_name))
                    print(f"Saved to not-like: {url}")
                else:
                    img.save(os.path.join(like_folder, clean_name))
                    print(f"Saved to like: {url}")
                
# Example usage
predict_and_save_images('/Users/nuttawatjanwisit/Desktop/nut/ml/data.txt', '/Users/nuttawatjanwisit/Desktop/nut/ml/test-like', '/Users/nuttawatjanwisit/Desktop/nut/ml/test-dot-like')
```
