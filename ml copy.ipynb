```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin

j = 84

def download_images(url, folder_name):
    # สร้างโฟลเดอร์ถ้ายังไม่มี
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    # ดึงเนื้อหาจากเว็บไซต์
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # ค้นหารูปภาพในหน้าเว็บ
    images = soup.find_all('img')

    # ดาวน์โหลดแต่ละรูปภาพ
    for i, image in enumerate(images):
        # แปลง relative URL เป็น absolute URL
        image_url = urljoin(url, image['src'])

        # ดึงข้อมูลรูปภาพ
        img_data = requests.get(image_url).content
        file_name = f"{folder_name}/image_{i+j}.jpg"

        with open(file_name, 'wb') as file:
            file.write(img_data)
            print(f"Downloaded {file_name}")
url = 'https://www.google.com/search?q=%E0%B9%81%E0%B8%A1%E0%B8%A7&tbm=isch&ved=2ahUKEwiGqNKey86DAxXSyDgGHZXPAjAQ2-cCegQIABAA&oq=%E0%B9%81%E0%B8%A1%E0%B8%A7&gs_lcp=CgNpbWcQARgAMgQIIxAnMgcIABCABBATMgcIABCABBATMgcIABCABBATMgcIABCABBATMgcIABCABBATMgcIABCABBATMgcIABCABBATMgcIABCABBATMgcIABCABBATOgcIIxDqAhAnOgUIABCABDoECAAQHlDiDFj4JGDWM2gBcAB4AYABiwGIAcQMkgEEMTYuM5gBAKABAaoBC2d3cy13aXotaW1nsAEKwAEB&sclient=img&ei=QFWcZYaIE9KR4-EPlZ-LgAM&bih=1180&biw=820'
# ใช้งานฟังก์ชัน
download_images(url, '/Users/nuttawatjanwisit/Desktop/nut/ml/ex-img2')
```

```python
import requests
from bs4 import BeautifulSoup
import re

# URL ของหน้าเว็บที่คุณต้องการสแคปปิ้ง
url = 'https://tinder.com/app/recs'

# ส่งคำขอไปยังหน้าเว็บ
response = requests.get(url)

# ตรวจสอบว่าคำขอเสร็จสมบูรณ์หรือไม่
if response.status_code == 200:
    # ใช้ BeautifulSoup ในการวิเคราะห์เนื้อหาของหน้าเว็บ
    soup = BeautifulSoup(response.content, 'html.parser')

    # ค้นหา <div> elements ที่มีคุณสมบัติตามที่กำหนด
    divs = soup.find_all('div', {'role': 'img', 'class': 'Bdrs(8px) Bgz(cv) Bgp(c) StretchedBox'})

    # วนลูปผ่านแต่ละ div และหาลิงก์
    for div in divs:
        style = div.get('style')
        if style:
            # ใช้ regular expression เพื่อหาลิงก์
            match = re.search(r'https://images-[^"]+', style)
            if match:
                print(match.group())
    print("not")
else:
    print("Failed to retrieve the webpage")
```

```python
import os

# ตั้งค่าตำแหน่งของโฟลเดอร์ที่มีไฟล์ที่คุณต้องการเปลี่ยนชื่อ
folder_path = '/Users/nuttawatjanwisit/Desktop/nut/ml/ex-img2'

# ดึงรายชื่อไฟล์ทั้งหมดในโฟลเดอร์
files = os.listdir(folder_path)

# วนลูปเพื่อเปลี่ยนชื่อไฟล์แต่ละไฟล์
for i, file in enumerate(files):
    # สร้างชื่อไฟล์ใหม่
    new_filename = f"image_{i}.jpg"

    # ตั้งค่าตำแหน่งเต็มของไฟล์เดิมและไฟล์ใหม่
    old_file = os.path.join(folder_path, file)
    new_file = os.path.join('/Users/nuttawatjanwisit/Desktop/nut/ml/ex-img3', new_filename)

    # เปลี่ยนชื่อไฟล์
    os.rename(old_file, new_file)

    print(f"Renamed '{file}' to '{new_filename}'")
```
